```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.txt', delimiter="\t", header=None, names=["N", "xa", "xp"])
```

```python
data
```

```python
data = data.sort_values('xp')
```

```python
data['xA'] =  np.nan
```

```python
data
```

```python
Xa = data['xa']
Xp = data['xp']
```

```python
Xa
```

```python
def consistency_regressive_sequence(finish_ind, len_seq):
    ind = finish_ind - len_seq - 1
    value = 0.0

    while ind < finish_ind + 1:
      value+=Xa[ind]
      #print(value)
      ind+=1

    value/=(len_seq + 2)
    #print(value)
    ind = finish_ind - len_seq - 1

    while ind < finish_ind + 1:
      data['xA'][ind] = value
      #print(data['xA'])
      ind+=1

    return data
```

```python
start_ind = 0
len_sequence = 0
data['xA'][0] = Xa[0]
i = 1

while i < len(Xa):
  #print(i, " ", start_ind, " ", Xa[i]," ", Xa[start_ind])
  if (Xa[i] >= Xa[start_ind]):
    if (len_sequence == 0):
      data['xA'][i] = Xa[i]
      start_ind +=1
    else:
      data = consistency_regressive_sequence(i,len_sequence)
      start_ind = start_ind + len_sequence + 1
      len_sequence=0
  else:
    len_sequence+=1
  i += 1
```

```python
data
```

```python
def consistency_same_sequence(finish_ind, len_seq):
    ind = finish_ind - len_seq - 1
    value = 0.0

    while ind < finish_ind:
      value+=Xa[ind]
      ind+=1

    value/=(len_seq+1)
    ind = finish_ind - len_seq - 1

    while ind < finish_ind:
      data['xA'][ind] = value
      ind+=1

    return data
```

```python
start_ind = 0
len_sequence = 0
i = 1

while i < len(Xp):
  #print(i, " ", start_ind, " ", Xp[i]," ", Xp[start_ind])
  if (Xp[i] != Xp[start_ind]):
    if(len_sequence != 0):
      data = consistency_same_sequence(i,len_sequence)
      start_ind+=len_sequence
      len_sequence=0
    start_ind+=1
  else:
    len_sequence+=1
  i += 1
```

```python
data
```

```python
average_xa = np.average(data['xa'])
average_xa
```

```python
average_xp = np.average(data['xp'])
average_xp
```

```python
average_xA = np.average(data['xA'])
average_xA
```

```python
average_xa_xA = np.average(data['xa']*data['xA'])
average_xa_xA
```

```python
average_xp_xA = np.average(data['xp']*data['xA'])
average_xp_xA
```

```python
D_xa = np.average((data['xa']-average_xa)**2)
D_xa
```

```python
D_xp = np.average((data['xp']-average_xp)**2)
D_xp
```

```python
D_xA = np.average((data['xA']-average_xA)**2)
D_xA
```

```python
r_xa_xA = (average_xa_xA - average_xa * average_xA) / (np.sqrt(D_xa*D_xA))
r_xa_xA
```

```python
r_xp_xA = (average_xp_xA - average_xp * average_xA) / (np.sqrt(D_xp*D_xA))
r_xp_xA
```

```python
d_a_A = 1 - r_xa_xA
d_a_A
```

```python
d_p_P = 1 - r_xp_xA
d_p_P
```

```python
d_a_p = np.average(d_p_P+d_a_A)
d_a_p
```
